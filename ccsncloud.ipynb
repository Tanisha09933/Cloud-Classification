```python
import tensorflow as tf
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.utils import class_weight
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras import layers, models

# GPU Check
print("GPU is", "available" if tf.config.list_physical_devices('GPU') else "not available")

# Config
image_height, image_width = 256, 256
batch_size = 32
dataset_path = '/kaggle/input/cirrus-cumulus-stratus-nimbus-ccsn-database/CCSN_v2'

# Load dataset (unshuffled first to compute class weights)
raw_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    dataset_path,
    image_size=(image_height, image_width),
    batch_size=batch_size,
    shuffle=False
)

class_names = raw_dataset.class_names
num_classes = len(class_names)

# Analyze Class Distribution
all_labels = []
for _, labels in raw_dataset:
    all_labels.extend(labels.numpy())

class_counts = np.bincount(all_labels)
print("\nClass Distribution:")
for i, count in enumerate(class_counts):
    print(f"{class_names[i]}: {count} images")

# Compute Class Weights
weights = class_weight.compute_class_weight(class_weight='balanced',
                                             classes=np.unique(all_labels),
                                             y=all_labels)
class_weights = {i: w for i, w in enumerate(weights)}
print("\nClass Weights:", class_weights)

# Reload with shuffle
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    dataset_path,
    image_size=(image_height, image_width),
    batch_size=batch_size,
    shuffle=True,
    seed=42
)

# Train/Val Split
val_split = 0.2
val_size = int(len(dataset) * val_split)
ds_train = dataset.skip(val_size)
ds_val = dataset.take(val_size)

# Gaussian Kernel
def get_gaussian_kernel(kernel_size=5, sigma=1.0):
    x = tf.range(-kernel_size // 2 + 1, kernel_size // 2 + 1, dtype=tf.float32)
    x = tf.math.exp(-(x**2) / (2 * sigma**2))
    kernel_1d = x / tf.reduce_sum(x)
    kernel_2d = tf.tensordot(kernel_1d, kernel_1d, axes=0)
    kernel_2d = kernel_2d[:, :, tf.newaxis, tf.newaxis]
    return tf.repeat(kernel_2d, 3, axis=2)

gaussian_kernel = get_gaussian_kernel()

# Sobel Filter
def get_sobel_kernels():
    sobel_x = tf.constant([[-1., 0., 1.],
                           [-2., 0., 2.],
                           [-1., 0., 1.]], dtype=tf.float32)
    sobel_y = tf.constant([[-1., -2., -1.],
                           [0., 0., 0.],
                           [1., 2., 1.]], dtype=tf.float32)
    sobel_x = sobel_x[:, :, tf.newaxis, tf.newaxis]
    sobel_y = sobel_y[:, :, tf.newaxis, tf.newaxis]
    return tf.repeat(sobel_x, 3, axis=2), tf.repeat(sobel_y, 3, axis=2)

sobel_x, sobel_y = get_sobel_kernels()

# Laplacian Filter
def get_laplacian_kernel():
    laplacian = tf.constant([[0., 1., 0.],
                             [1., -4., 1.],
                             [0., 1., 0.]], dtype=tf.float32)
    laplacian = laplacian[:, :, tf.newaxis, tf.newaxis]
    return tf.repeat(laplacian, 3, axis=2)

laplacian_kernel = get_laplacian_kernel()

# Combined Preprocessing Function
def preprocess(images, labels):
    images = tf.cast(images, tf.float32) / 255.0
    blurred = tf.nn.depthwise_conv2d(images, gaussian_kernel, strides=[1, 1, 1, 1], padding='SAME')
    edges_x = tf.nn.depthwise_conv2d(blurred, sobel_x, strides=[1, 1, 1, 1], padding='SAME')
    edges_y = tf.nn.depthwise_conv2d(blurred, sobel_y, strides=[1, 1, 1, 1], padding='SAME')
    sobel_edges = tf.sqrt(tf.square(edges_x) + tf.square(edges_y))
    laplacian_edges = tf.nn.depthwise_conv2d(blurred, laplacian_kernel, strides=[1, 1, 1, 1], padding='SAME')
    enhanced = blurred + sobel_edges + laplacian_edges
    images = tf.clip_by_value(enhanced, 0.0, 1.0)
    return images, labels

ds_train = ds_train.map(preprocess).prefetch(tf.data.AUTOTUNE)
ds_val = ds_val.map(preprocess).prefetch(tf.data.AUTOTUNE)

# Block Attention Module
class BlockAttentionModule(layers.Layer):
    def __init__(self, channels, reduction_ratio=16):
        super(BlockAttentionModule, self).__init__()
        self.channels = channels
        self.reduction_ratio = reduction_ratio

        self.avg_pool = layers.GlobalAveragePooling2D()
        self.fc1 = layers.Dense(channels // reduction_ratio, activation='relu')
        self.fc2 = layers.Dense(channels)
        self.conv = layers.Conv2D(1, kernel_size=7, padding='same', activation='sigmoid')

    def call(self, inputs):
        avg = self.avg_pool(inputs)
        avg = self.fc1(avg)
        avg = self.fc2(avg)
        channel_att = tf.nn.sigmoid(avg)
        channel_att = tf.reshape(channel_att, (-1, 1, 1, self.channels))
        channel_refined = inputs * channel_att

        spatial_att = self.conv(channel_refined)
        spatial_refined = channel_refined * spatial_att

        return channel_refined + spatial_refined

# CNN Model with BAM
model = models.Sequential([
    layers.Conv2D(32, 3, padding='same', activation='mish', input_shape=(image_height, image_width, 3)),
    BlockAttentionModule(32),
    layers.MaxPooling2D(),

    layers.Conv2D(64, 3, padding='same', activation='mish'),
    BlockAttentionModule(64),
    layers.MaxPooling2D(),

    layers.Conv2D(128, 3, padding='same', activation='mish'),
    BlockAttentionModule(128),
    layers.MaxPooling2D(),

    layers.Dropout(0.3),
    layers.Flatten(),
    layers.Dense(128, activation='mish'),
    layers.Dense(num_classes, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(ds_train, validation_data=ds_val, epochs=50, class_weight=class_weights)

test_loss, test_acc = model.evaluate(ds_val)
print("\nTest accuracy:", test_acc)

true_labels, predicted_labels = [], []
for images, labels in ds_val:
    preds = model.predict(images)
    predicted_labels.extend(np.argmax(preds, axis=1))
    true_labels.extend(labels.numpy())

cm = confusion_matrix(true_labels, predicted_labels)
cm_df = pd.DataFrame(cm, index=class_names, columns=class_names)

plt.figure(figsize=(6, 5))
sns.heatmap(cm_df, annot=True, cmap='Blues', fmt='g')
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()

report = classification_report(true_labels, predicted_labels, target_names=class_names)
print("\nClassification Report:\n", report)
```

```python
# Cell 1: Training History Plots (Accuracy and Loss Curves)
import matplotlib.pyplot as plt
import numpy as np
import warnings

# Suppress the specific matplotlib warning
warnings.filterwarnings('ignore', category=RuntimeWarning, message='invalid value encountered in less')

# Plot training history
plt.figure(figsize=(15, 5))

# Accuracy plot
plt.subplot(1, 3, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='s')
plt.title('Model Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

# Loss plot
plt.subplot(1, 3, 2)
plt.plot(history.history['loss'], label='Training Loss', marker='o')
plt.plot(history.history['val_loss'], label='Validation Loss', marker='s')
plt.title('Model Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

# Combined plot
plt.subplot(1, 3, 3)
plt.plot(history.history['accuracy'], label='Train Acc', alpha=0.7)
plt.plot(history.history['val_accuracy'], label='Val Acc', alpha=0.7)
plt.plot(history.history['loss'], label='Train Loss', alpha=0.7)
plt.plot(history.history['val_loss'], label='Val Loss', alpha=0.7)
plt.title('Training Overview')
plt.xlabel('Epoch')
plt.ylabel('Metric Value')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print final metrics
print(f"Final Training Accuracy: {history.history['accuracy'][-1]:.4f}")
print(f"Final Validation Accuracy: {history.history['val_accuracy'][-1]:.4f}")
print(f"Final Training Loss: {history.history['loss'][-1]:.4f}")
print(f"Final Validation Loss: {history.history['val_loss'][-1]:.4f}")

# Cell 2: ROC-AUC Analysis with Fixed Color Generation
from sklearn.metrics import roc_curve, auc, roc_auc_score
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
import numpy as np

# Robust color generation function
def get_robust_colors(n_colors):
    """Generate robust colors without causing matplotlib warnings"""
    if n_colors <= 10:
        # Use discrete colors from tab10
        indices = np.arange(n_colors)
        return plt.cm.tab10(indices)
    elif n_colors <= 20:
        # Use discrete colors from tab20
        indices = np.arange(n_colors)
        return plt.cm.tab20(indices)
    else:
        # For more than 20 colors, use continuous colormap
        return plt.cm.viridis(np.linspace(0, 1, n_colors))

# Get predicted probabilities for ROC analysis
all_probs = []
all_true_labels = []

for images, labels in ds_val:
    probs = model.predict(images, verbose=0)
    all_probs.extend(probs)
    all_true_labels.extend(labels.numpy())

all_probs = np.array(all_probs)
all_true_labels = np.array(all_true_labels)

# Data validation and cleaning
print(f"Checking data integrity...")
print(f"All probs shape: {all_probs.shape}")
print(f"NaN in probs: {np.isnan(all_probs).any()}")
print(f"Inf in probs: {np.isinf(all_probs).any()}")

# Clean invalid values
if np.isnan(all_probs).any() or np.isinf(all_probs).any():
    print("Cleaning invalid values in probabilities...")
    all_probs = np.nan_to_num(all_probs, nan=0.0, posinf=1.0, neginf=0.0)

# Ensure probabilities are in valid range [0, 1]
all_probs = np.clip(all_probs, 0.0, 1.0)

# Binarize labels for multiclass ROC
y_test_bin = label_binarize(all_true_labels, classes=range(num_classes))

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

plt.figure(figsize=(15, 10))

# Get robust colors for 11 cloud types
colors = get_robust_colors(num_classes)

# Plot ROC curves for each class
plt.subplot(2, 2, 1)
for i in range(num_classes):
    try:
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], all_probs[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
        plt.plot(fpr[i], tpr[i], color=colors[i], lw=2,
                 label=f'{class_names[i]} (AUC = {roc_auc[i]:.3f})')
    except Exception as e:
        print(f"Error plotting ROC for class {i}: {e}")
        roc_auc[i] = 0.0

plt.plot([0, 1], [0, 1], 'k--', lw=2, alpha=0.5)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Each Cloud Type')
plt.legend(loc="lower right", fontsize=8)
plt.grid(True, alpha=0.3)

# Compute macro-average ROC curve
try:
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(num_classes) if i in fpr]))
    mean_tpr = np.zeros_like(all_fpr)
    valid_classes = 0
    
    for i in range(num_classes):
        if i in fpr and len(fpr[i]) > 0:
            mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
            valid_classes += 1
    
    if valid_classes > 0:
        mean_tpr /= valid_classes
        fpr["macro"] = all_fpr
        tpr["macro"] = mean_tpr
        roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])
    else:
        roc_auc["macro"] = 0.0
except Exception as e:
    print(f"Error computing macro-average ROC: {e}")
    roc_auc["macro"] = 0.0

# Plot macro-average ROC curve
plt.subplot(2, 2, 2)
if "macro" in fpr and "macro" in tpr:
    plt.plot(fpr["macro"], tpr["macro"], color='navy', linestyle=':', linewidth=4,
             label=f'Macro-average ROC (AUC = {roc_auc["macro"]:.3f})')
plt.plot([0, 1], [0, 1], 'k--', lw=2, alpha=0.5)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Macro-Average ROC Curve')
plt.legend(loc="lower right")
plt.grid(True, alpha=0.3)

# AUC scores bar plot
plt.subplot(2, 2, 3)
auc_scores = [roc_auc.get(i, 0.0) for i in range(num_classes)]
bars = plt.bar(class_names, auc_scores, color=colors, alpha=0.7)
plt.title('AUC Scores by Cloud Type')
plt.ylabel('AUC Score')
plt.xticks(rotation=45, ha='right')
plt.ylim([0, 1])

# Add value labels on bars
for bar, score in zip(bars, auc_scores):
    if score > 0:
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                 f'{score:.3f}', ha='center', va='bottom', fontsize=8)

plt.grid(True, alpha=0.3)

# Prediction confidence distribution
plt.subplot(2, 2, 4)
max_probs = np.max(all_probs, axis=1)
# Clean max_probs to avoid any remaining invalid values
max_probs = np.nan_to_num(max_probs, nan=0.0)
max_probs = np.clip(max_probs, 0.0, 1.0)

plt.hist(max_probs, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Prediction Confidence Distribution')
plt.xlabel('Max Probability')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print AUC summary
print("\nAUC Scores Summary for Cloud Types:")
print("-" * 40)
for i in range(num_classes):
    print(f"{class_names[i]}: {roc_auc.get(i, 0.0):.4f}")
print(f"Macro-Average: {roc_auc.get('macro', 0.0):.4f}")

# Calculate and print overall multiclass AUC
try:
    overall_auc = roc_auc_score(all_true_labels, all_probs, multi_class='ovr', average='macro')
    print(f"Overall Multiclass AUC: {overall_auc:.4f}")
except Exception as e:
    print(f"Overall Multiclass AUC: Could not compute - {e}")

# Cell 3: Enhanced Correlation Analysis
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix

# Create feature correlation analysis with error handling
def extract_features_from_model(model, data, max_batches=2):
    """Extract features from intermediate layers with robust error handling"""
    try:
        print("Starting feature extraction...")
        
        # Create a model that outputs features from different layers
        layer_outputs = []
        layer_names = []
        
        for i, layer in enumerate(model.layers):
            if isinstance(layer, (tf.keras.layers.Conv2D, tf.keras.layers.Dense)):
                try:
                    if hasattr(layer, 'output_shape') and len(layer.output_shape) > 1:
                        layer_outputs.append(layer.output)
                        layer_names.append(f"Layer_{i}_{layer.name}")
                except Exception as e:
                    print(f"Skipping layer {i}: {e}")
                    continue
        
        if not layer_outputs:
            print("No suitable layers found for feature extraction")
            return None, None, None
            
        feature_extractor = tf.keras.Model(inputs=model.input, outputs=layer_outputs)
        
        # Extract features from a small sample of validation data
        sample_features = []
        sample_labels = []
        
        batch_count = 0
        for images, labels in data.take(max_batches):
            try:
                features = feature_extractor(images, training=False)
                sample_labels.extend(labels.numpy())
                
                # Flatten and aggregate features from each layer
                batch_features = []
                for layer_feat in features:
                    layer_feat_clean = tf.where(tf.math.is_finite(layer_feat), layer_feat, 0.0)
                    
                    if len(layer_feat_clean.shape) == 4:  # Conv layer
                        # Global average pooling for conv features
                        pooled = tf.reduce_mean(layer_feat_clean, axis=[1, 2])
                        batch_features.append(pooled.numpy())
                    else:  # Dense layer
                        batch_features.append(layer_feat_clean.numpy())
                
                # Concatenate features from all layers for this batch
                if batch_features:
                    concatenated = np.concatenate(batch_features, axis=1)
                    # Clean concatenated features
                    concatenated = np.nan_to_num(concatenated, nan=0.0, posinf=1.0, neginf=-1.0)
                    sample_features.extend(concatenated)
                    
                batch_count += 1
                print(f"Processed batch {batch_count}/{max_batches}")
                
            except Exception as e:
                print(f"Error processing batch {batch_count + 1}: {e}")
                continue
        
        if sample_features:
            features_array = np.array(sample_features)
            labels_array = np.array(sample_labels)
            print(f"Successfully extracted features: {features_array.shape}")
            return features_array, labels_array, layer_names
        else:
            print("No features could be extracted")
            return None, None, None
            
    except Exception as e:
        print(f"Feature extraction failed: {e}")
        return None, None, None

# Extract features for correlation analysis
print("Extracting features for correlation analysis...")
features, labels, layer_names = extract_features_from_model(model, ds_val, max_batches=2)

plt.figure(figsize=(20, 12))

if features is not None and not np.isnan(features).any() and not np.isinf(features).any():
    print(f"Using extracted features: {features.shape}")
    
    # 1. Class-wise feature correlation
    plt.subplot(2, 3, 1)
    
    # Create a DataFrame with selected features and labels
    n_features = min(50, features.shape[1])  # Use first 50 features or all if less
    feature_df = pd.DataFrame(features[:, :n_features])
    feature_df['class'] = labels
    
    # Compute correlation between features and class labels
    correlations = []
    for col in feature_df.columns[:-1]:
        try:
            corr = feature_df[col].corr(feature_df['class'])
            correlations.append(corr if not np.isnan(corr) else 0.0)
        except:
            correlations.append(0.0)
    
    # Plot feature-class correlations
    plt.bar(range(len(correlations)), correlations, alpha=0.7, color='steelblue')
    plt.title('Feature-Class Correlations')
    plt.xlabel('Feature Index')
    plt.ylabel('Correlation with Class')
    plt.grid(True, alpha=0.3)

    # 2. Feature correlation heatmap (subset)
    plt.subplot(2, 3, 2)
    n_heatmap_features = min(20, features.shape[1])
    feature_subset = features[:, :n_heatmap_features]
    
    try:
        corr_matrix = np.corrcoef(feature_subset.T)
        # Clean correlation matrix
        corr_matrix = np.nan_to_num(corr_matrix, nan=0.0)
        sns.heatmap(corr_matrix, cmap='coolwarm', center=0, 
                    square=True, annot=False, cbar_kws={"shrink": .8})
        plt.title(f'Feature Correlation Heatmap (First {n_heatmap_features} Features)')
    except Exception as e:
        print(f"Error creating correlation heatmap: {e}")
        plt.text(0.5, 0.5, 'Correlation heatmap\nunavailable', 
                ha='center', va='center', transform=plt.gca().transAxes)
        plt.title('Feature Correlation Heatmap')

else:
    print("Using alternative analysis due to feature extraction issues...")
    
    # Alternative analysis when feature extraction fails
    # 1. Class distribution
    plt.subplot(2, 3, 1)
    bars = plt.bar(class_names, class_counts, alpha=0.7, color='skyblue')
    plt.title('Cloud Type Distribution')
    plt.ylabel('Number of Samples')
    plt.xticks(rotation=45, ha='right')
    for i, count in enumerate(class_counts):
        if count > 0:
            plt.text(i, count + max(class_counts)*0.01, str(count), 
                    ha='center', va='bottom', fontsize=8)
    plt.grid(True, alpha=0.3)
    
    # 2. Performance correlation matrix
    plt.subplot(2, 3, 2)
    # Create correlation matrix from available metrics
    metrics_data = []
    for i in range(num_classes):
        class_mask = np.array(true_labels) == i
        if np.sum(class_mask) > 0:
            class_pred = np.array(predicted_labels)[class_mask]
            class_acc = np.mean(class_pred == i)
            class_precision = np.sum((np.array(predicted_labels) == i) & (np.array(true_labels) == i)) / max(1, np.sum(np.array(predicted_labels) == i))
            class_recall = np.sum((np.array(predicted_labels) == i) & (np.array(true_labels) == i)) / max(1, np.sum(np.array(true_labels) == i))
            metrics_data.append([class_counts[i], class_acc, class_precision, class_recall])
        else:
            metrics_data.append([0, 0, 0, 0])
    
    metrics_df = pd.DataFrame(metrics_data, 
                             columns=['Sample_Count', 'Accuracy', 'Precision', 'Recall'],
                             index=class_names)
    
    try:
        corr_matrix = metrics_df.corr()
        # Clean correlation matrix
        corr_matrix = corr_matrix.fillna(0)
        sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, 
                    square=True, cbar_kws={"shrink": .8}, fmt='.2f')
        plt.title('Performance Metrics Correlation')
    except Exception as e:
        print(f"Error creating performance correlation: {e}")
        plt.text(0.5, 0.5, 'Performance correlation\nunavailable', 
                ha='center', va='center', transform=plt.gca().transAxes)

# Rest of the visualization code continues with the same error handling pattern...
# (The remaining subplots follow the same robust approach)

# 3. Per-class accuracy
plt.subplot(2, 3, 3)
class_accuracy = []
for i in range(num_classes):
    class_mask = np.array(true_labels) == i
    if np.sum(class_mask) > 0:
        class_pred = np.array(predicted_labels)[class_mask]
        class_acc = np.mean(class_pred == i)
        class_accuracy.append(class_acc)
    else:
        class_accuracy.append(0)

colors_subset = get_robust_colors(num_classes)
bars = plt.bar(class_names, class_accuracy, alpha=0.7, color=colors_subset)
plt.title('Per-Class Accuracy')
plt.ylabel('Accuracy')
plt.xticks(rotation=45, ha='right')
plt.ylim([0, 1])

# Add value labels on bars
for bar, acc in zip(bars, class_accuracy):
    if acc > 0:
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                 f'{acc:.3f}', ha='center', va='bottom', fontsize=8)

plt.grid(True, alpha=0.3)

# 4. Prediction confidence by class
plt.subplot(2, 3, 4)
confidence_by_class = []
for i in range(num_classes):
    class_mask = np.array(true_labels) == i
    if np.sum(class_mask) > 0:
        class_probs = all_probs[class_mask]
        class_confidence = np.max(class_probs, axis=1)
        # Clean confidence values
        class_confidence = np.nan_to_num(class_confidence, nan=0.0)
        confidence_by_class.append(np.mean(class_confidence))
    else:
        confidence_by_class.append(0)

plt.bar(class_names, confidence_by_class, alpha=0.7, color=colors_subset)
plt.title('Average Prediction Confidence by Cloud Type')
plt.ylabel('Average Confidence')
plt.xticks(rotation=45, ha='right')
plt.ylim([0, 1])

# Add value labels
for i, conf in enumerate(confidence_by_class):
    if conf > 0:
        plt.text(i, conf + 0.01, f'{conf:.3f}', ha='center', va='bottom', fontsize=8)

plt.grid(True, alpha=0.3)

# 5. Class size vs Performance correlation
plt.subplot(2, 3, 5)
plt.scatter(class_counts, class_accuracy, s=100, alpha=0.7, color='purple')
for i, (count, acc) in enumerate(zip(class_counts, class_accuracy)):
    if count > 0 and acc > 0:
        plt.annotate(class_names[i][:3], (count, acc), xytext=(5, 5), 
                    textcoords='offset points', fontsize=8)

plt.xlabel('Number of Training Samples')
plt.ylabel('Class Accuracy')
plt.title('Class Size vs Performance')
plt.grid(True, alpha=0.3)

# Calculate and show correlation
try:
    size_perf_corr = np.corrcoef(class_counts, class_accuracy)[0, 1]
    if np.isnan(size_perf_corr):
        size_perf_corr = 0.0
except:
    size_perf_corr = 0.0

plt.text(0.05, 0.95, f'Correlation: {size_perf_corr:.3f}', 
         transform=plt.gca().transAxes, bbox=dict(boxstyle="round", facecolor='white', alpha=0.8))

# 6. Normalized confusion matrix
plt.subplot(2, 3, 6)
try:
    confusion_matrix_data = confusion_matrix(true_labels, predicted_labels)
    # Normalize confusion matrix
    cm_normalized = confusion_matrix_data.astype('float') / (confusion_matrix_data.sum(axis=1)[:, np.newaxis] + 1e-8)
    # Clean normalized matrix
    cm_normalized = np.nan_to_num(cm_normalized, nan=0.0)
    
    sns.heatmap(cm_normalized, annot=True, fmt='.2f', cmap='Blues',
                xticklabels=[name[:3] for name in class_names], 
                yticklabels=[name[:3] for name in class_names])
    plt.title('Normalized Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
except Exception as e:
    print(f"Error creating confusion matrix: {e}")
    plt.text(0.5, 0.5, 'Confusion matrix\nunavailable', 
            ha='center', va='center', transform=plt.gca().transAxes)

plt.tight_layout()
plt.show()

# Print correlation analysis summary
print("\nCorrelation Analysis Summary:")
print("-" * 40)
print(f"Class Size vs Performance Correlation: {size_perf_corr:.4f}")
print(f"Mean Prediction Confidence: {np.mean(max_probs):.4f}")
print(f"Std Prediction Confidence: {np.std(max_probs):.4f}")

# Per-class statistics
print("\nPer-Class Statistics for Cloud Types:")
print("-" * 50)
for i in range(num_classes):
    print(f"{class_names[i]}:")
    print(f"  Samples: {class_counts[i]}")
    print(f"  Accuracy: {class_accuracy[i]:.4f}")
    print(f"  Avg Confidence: {confidence_by_class[i]:.4f}")
    print(f"  AUC Score: {roc_auc.get(i, 0.0):.4f}")
    print()

# Cell 4: Additional Model Performance Metrics with Enhanced Error Handling
from sklearn.metrics import precision_recall_curve, average_precision_score
import matplotlib.pyplot as plt

# Precision-Recall Curves
plt.figure(figsize=(15, 10))

# Plot precision-recall curves for each class
plt.subplot(2, 2, 1)
colors = get_robust_colors(num_classes)

for i in range(num_classes):
    try:
        if np.sum(y_test_bin[:, i]) > 0:  # Only if class has positive samples
            precision, recall, _ = precision_recall_curve(y_test_bin[:, i], all_probs[:, i])
            avg_precision = average_precision_score(y_test_bin[:, i], all_probs[:, i])
            
            # Clean precision and recall values
            precision = np.nan_to_num(precision, nan=0.0)
            recall = np.nan_to_num(recall, nan=0.0)
            
            plt.plot(recall, precision, color=colors[i], lw=2,
                     label=f'{class_names[i][:8]} (AP = {avg_precision:.3f})')
    except Exception as e:
        print(f"Error plotting PR curve for class {i}: {e}")

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curves')
plt.legend(loc="lower left", fontsize=8)
plt.grid(True, alpha=0.3)

# Model calibration plot
plt.subplot(2, 2, 2)
try:
    from sklearn.calibration import calibration_curve
    
    # Plot calibration curve for each class
    for i in range(min(5, num_classes)):  # Limit to first 5 classes for readability
        if np.sum(y_test_bin[:, i]) > 0:  # Only if class has positive samples
            try:
                fraction_of_positives, mean_predicted_value = calibration_curve(
                    y_test_bin[:, i], all_probs[:, i], n_bins=10)
                
                # Clean calibration values
                fraction_of_positives = np.nan_to_num(fraction_of_positives, nan=0.0)
                mean_predicted_value = np.nan_to_num(mean_predicted_value, nan=0.0)
                
                plt.plot(mean_predicted_value, fraction_of_positives, "s-",
                        label=f'{class_names[i][:8]}', color=colors[i])
            except Exception as e:
                print(f"Error in calibration curve for class {i}: {e}")

    plt.plot([0, 1], [0, 1], "k:", label="Perfectly calibrated")
    plt.xlabel('Mean Predicted Probability')
    plt.ylabel('Fraction of Positives')
    plt.title('Calibration Plot (First 5 Classes)')
    plt.legend(fontsize=8)
    plt.grid(True, alpha=0.3)
except ImportError:
    plt.text(0.5, 0.5, 'Calibration plot\nrequires sklearn', 
            ha='center', va='center', transform=plt.gca().transAxes)
    plt.title('Calibration Plot')

# Learning curves
plt.subplot(2, 2, 3)
epochs = range(1, len(history.history['accuracy']) + 1)
train_acc = np.array(history.history['accuracy'])
val_acc = np.array(history.history['val_accuracy'])

# Clean accuracy values
train_acc = np.nan_to_num(train_acc, nan=0.0)
val_acc = np.nan_to_num(val_acc, nan=0.0)

plt.plot(epochs, train_acc, 'b-', label='Training Accuracy', linewidth=2)
plt.plot(epochs, val_acc, 'r-', label='Validation Accuracy', linewidth=2)
plt.fill_between(epochs, train_acc, val_acc, alpha=0.3)
plt.title('Training vs Validation Accuracy Gap')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

# Model complexity vs accuracy
plt.subplot(2, 2, 4)
# Calculate model complexity metrics
total_params = model.count_params()
trainable_params = sum([tf.keras.backend.count_params(w) for w in model.trainable_weights])

complexity_metrics = {
    'Total Params': total_params,
    'Trainable Params': trainable_params,
    'Model Layers': len(model.layers),
    'Conv Layers': sum([1 for layer in model.layers if isinstance(layer, tf.keras.layers.Conv2D)]),
    'Dense Layers': sum([1 for layer in model.layers if isinstance(layer, tf.keras.layers.Dense)])
}

bars = plt.bar(range(len(complexity_metrics)), list(complexity_metrics.values()), 
               color='lightblue', alpha=0.7)
plt.xticks(range(len(complexity_metrics)), 
           [key.replace(' ', '\n') for key in complexity_metrics.keys()], 
           rotation=0, fontsize=8)
plt.title('Model Complexity Metrics')
plt.ylabel('Count')
plt.yscale('log')  # Log scale due to large parameter counts

for i, (key, value) in enumerate(complexity_metrics.items()):
    plt.text(i, value, f'{value:,}' if value > 1000 else str(value), 
            ha='center', va='bottom', fontsize=8)

plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nModel Complexity Summary:")
print(f"Total Parameters: {total_params:,}")
print(f"Trainable Parameters: {trainable_params:,}")
print(f"Final Validation Accuracy: {val_acc[-1]:.4f}")
if val_acc[-1] > 0:
    print(f"Parameters per Accuracy Point: {total_params/val_acc[-1]:,.0f}")
```
